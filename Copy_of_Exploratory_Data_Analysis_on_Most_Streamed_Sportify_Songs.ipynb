```python
import pandas as pd

# Load the dataset
file_path = '/content/Most Streamed Spotify Songs 2024.csv'
df = pd.read_csv(file_path, encoding='latin-1')

# Convert 'Release Date' to datetime format
df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')

# Ensure 'Spotify Streams' is in a numeric format (removing commas)
df['Spotify Streams'] = df['Spotify Streams'].astype(str).str.replace(',', '')
df['Spotify Streams'] = pd.to_numeric(df['Spotify Streams'], errors='coerce').fillna(0).astype(int)

# Filter the dataframe for songs released in 2024
tracks_2024 = df[df['Release Date'].dt.year == 2024]

# Group by artist and sum the Spotify streams
artist_total_streams_2024 = tracks_2024.groupby('Artist')['Spotify Streams'].sum().reset_index()

# Sort by Spotify streams in descending order and get the top 5 artists
top_5_artists_2024 = artist_total_streams_2024.sort_values(by='Spotify Streams', ascending=False).head(5)

# Print the top 5 artists
print("Top 5 artists in 2024 based on Spotify streams:")
print(top_5_artists_2024)
```

```python
import pandas as pd

# Load the dataset
file_path = '/content/Most Streamed Spotify Songs 2024.csv'
df = pd.read_csv(file_path, encoding='latin-1')

# Convert 'Release Date' to datetime format
df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')

# Ensure 'Spotify Streams' is in a numeric format (removing commas)
df['Spotify Streams'] = df['Spotify Streams'].astype(str).str.replace(',', '')
df['Spotify Streams'] = pd.to_numeric(df['Spotify Streams'], errors='coerce').fillna(0).astype(int)

# Extract the release year from 'Release Date'
df['Release Year'] = df['Release Date'].dt.year

# Group the dataframe by Year and sum the Spotify Streams
yearly_streams = df.groupby('Release Year')['Spotify Streams'].sum()

# Sort the yearly streams in descending order and get the top 5 years
top_5_years = yearly_streams.sort_values(ascending=False).head(5)

print("Top 5 years by total Spotify streams:")
print(top_5_years)
```

. What is the distribution of Spotify streams for the top 100 most streamed songs?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/Most Streamed Spotify Songs 2024.csv'
df = pd.read_csv(file_path, encoding='latin-1')

# Ensure 'Spotify Streams' is in a numeric format (removing commas)
df['Spotify Streams'] = df['Spotify Streams'].astype(str).str.replace(',', '')
df['Spotify Streams'] = pd.to_numeric(df['Spotify Streams'], errors='coerce').fillna(0).astype(int)

# Plot the distribution of Spotify streams
plt.figure(figsize=(10, 6))
plt.hist(df['Spotify Streams'], bins=20, edgecolor='k', alpha=0.7)
plt.title('Distribution of Spotify Streams')
plt.xlabel('Spotify Streams')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

 What is the trend of the number of songs released per year?


```python
# Convert 'Release Date' to datetime format
df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')

# Extract the release year
df['Release Year'] = df['Release Date'].dt.year

# Plot the number of songs released per year
songs_per_year = df['Release Year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
songs_per_year.plot(kind='bar')
plt.title('Number of Songs Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.grid(True)
plt.show()
```
